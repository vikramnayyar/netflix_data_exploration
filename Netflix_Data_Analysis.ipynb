## #3 Top Directors of Movies and TV Shows

#### #3.1 Exploding Director Column

```python
def explode_df(col):
    exploded_df = df[col].str.split(',').explode()
    df1 = df.drop([col], axis=1)
    df_dir = pd.concat([df1, exploded_df], axis=1)
    df_dir[col] = df_dir[col].str.strip()
    return df_dir
```

```python
df_dir = explode_df('director')
print(df_dir.shape, df.shape)
df_dir[df_dir['director'].str.find(',') != -1]
```

* All nested Directors have been exploded

#### #3.2 Familiarizing Director Column

```python
dir_counts = df_dir['director'].value_counts()
dir_counts = dir_counts[dir_counts.index != 'Not Available']
dir_counts
```

```python
plt.figure(figsize=(15,7))
dir_counts = dir_counts[:30]

df_dir_1 = df_dir[df_dir['director'].isin(dir_counts.index)]

sns.countplot(x='director', data = df_dir_1, hue='type', order=dir_counts.index, palette= custom_palette) #, order=dir_counts.index
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.xlabel(None)
plt.ylabel(None)
plt.legend(['Movies', 'TV Shows'])
plt.title('Top 30 Directors based on Total Movies and TV Shows')
plt.show()
```

```python
df = df[df['release_year'] >= 1992]
df['year_range']  = df['release_year']

def encode_yr(x):
    if x >=1992 and x < 2002:
        return '1992-2001'
    if x >= 2002 and x < 2012:
        return '2002-2011'
    if x >= 2012:
        return '2012-2021'

df['year_range'] = df['year_range'].apply(encode_yr)

df.head()
```

## #9. Preferred Duration of Movies & TV Shows

```python
df[df['type']=='Movie']['duration'].value_counts()[:40]
# print(df[df['type']=='Movie']['duration'].min(), df[df['type']=='Movie']['duration'].max())
```

```python
df_mov = df.loc[df['type']=='Movie']
df_mov['duration_bins'] = df_mov['duration'].str.split(' ').str[0]
df_mov['duration_bins'] = df_mov['duration_bins'].astype('int')

print(df_mov['duration_bins'].max(), df_mov['duration_bins'].min())

df_mov['duration_bins'] = df_mov['duration_bins'].apply(
                                        lambda x: '1 hour' if (x >= 40 and x <=80)
                                                            else 'short' if (x <= 30)
                                                            else '2 hour' if (x > 30 and x <=150)
                                                            else '3 hour' if (x > 150 and x <=210)
                                                            else '4 hour' if (x > 210)
                                                            else x)
```

```python
mov_duration_counts = df_mov['duration_bins'].value_counts()
mov_duration_counts = mov_duration_counts[:4]

show_duration_counts = df[df['type']=='TV Show']['duration'].value_counts()
show_duration_counts = show_duration_counts[:5]


plt.figure(figsize=(16,16))

plt.subplots_adjust(top=1.35)
plt.suptitle('Preferred Durations of Movies and TV Shows', fontsize=14.5)


plt.subplot(1,2,1)
plt.pie(mov_duration_counts, labels=mov_duration_counts.index, autopct='%.0f%%', startangle=90)
plt.title('Movies', fontsize=14)

plt.subplot(1,2,2)
plt.pie(show_duration_counts, labels=show_duration_counts.index, autopct='%.0f%%', startangle=90)
plt.title('TV Shows', fontsize=14)


plt.show()
```

* Almost **80%** of Movies are of **2 hours** duration
* Netflix should prefer movies of this duration as this is as per expectation of audience, longer movies may tend to loose interest of viewers.

* **70%** of TV Shows have just 1 seasons
* This indicates viewers dont prefer more seasons of a same TV Show, as this turns repeatitiveness in plot.
* Only few TV Shows less than **15%** should be considered for 2 seasons, that are highly expected to be favorite.
